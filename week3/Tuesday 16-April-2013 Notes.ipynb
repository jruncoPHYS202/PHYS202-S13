create a very simple plot

```python
import numpy as np
t = arange(0.0, 1.0+0.01, 0.01)
c = cos(2*2*pi*t)
s = sin(2*2*pi*t)
```

```python
plot(t, c, linewidth = 2.5, label = "$cos(4\pi*t)$", linestyle = ":") #can change line thicknes/ line style
plot(t, s, color = "red", label = "$sin(4\pi*t)$") #can change color
xlabel("time (s)")
ylabel("voltage (mV)")
# add title for convience
title("plot", fontsize = 30)
grid(True) #adds gridlines
savefig("simple_plot.pdf") #save plot
legend(loc = "lower left") #needed for label above
```
