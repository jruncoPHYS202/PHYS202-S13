```python
f = lambda x: (x+1)/(x-1)
g = lambda x: exp(-x)
x = arange(-4,4,.1)
plot(x,f(x), color = "red", label = '(t+1)/(t-1)')
plot(x,g(x), linestyle = ':', color = "blue", label = "exp(-t)")
xlabel("time (seconds)")
ylabel("f(t),g(t)")
legend(loc = "lower left")
ylim(-10,10)
annotate('discontinuous here', xy=(2.5,f(2.5)), xytext=(1, -3.),)
show()
```
