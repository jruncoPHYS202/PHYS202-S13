```python
f = lambda x: x**4 - 2*x + 1
```

```python
N = 100
a = 0.0
b = 2.0
h = (b-a)/N
s = .5*f(a) + .5*f(b)
for k in range(1,N):
    s += f(a+k*h)
print (h*s)
```

```python
N = 1000 #trap rule percent error
a = 0.0
b = 2.0
h = (b-a)/N
s = .5*f(a) + .5*f(b)
for k in range(1,N):
    s += f(a+k*h)
print (h*s)
print ((h*s)-4.4)/4.4 *100
```

The percent error using the trap rule at N=1000 is much greater than the percent error of using the simpson's rule with N=1000.  The trap rule is still accurate within 3 significant figures however, so it is still useful.
