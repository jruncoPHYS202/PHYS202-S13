The series, 1^1 + 2^2 + 3^3 + ... + 10^10 = 10405071317.

Find the last ten digits of the series, 1^1 + 2^2 + 3^3 + ... + 1000^1000.

```python
1000**1000
```

```python
def self_powers(x):
    """Finds the product of a self powers series"""
    y = 0
    for i in range(1,x+1): #loops through given range and sums up the self power series
        y += i**i
    return y
```

```python
print self_powers(3)
```

```python
print self_powers(10)
```

```python
print self_powers(1000)
```

As we can see, this definition works to find the self powers sum, so now let's find the length of the last 10 digits

```python
z = str(self_powers(10)) #turn it into a string
print len(z) #print length
```

```python
z1 = z[-10:] # get last 10 digits
print z1
```

This correctly returns the last 10 digits of the sum given in the problem, so now let's get the answer from self_powers(1000)
