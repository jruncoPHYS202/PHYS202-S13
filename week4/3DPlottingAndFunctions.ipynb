Electric dipole potential

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^{2} + (y-d)^{2}}} - \frac{kq}{\sqrt{x^{2} + (y+d)^{2}}}
\end{equation}

```python
#make the x-y grid to cover the plane -2<x<2 and -2<y<2
dx = .06
dy = .06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
x.ndim
y.shape
```

```python
k = 1; q = 1; d = 1
V = (k*q/(x**2 + (y - d)**2)**(1/2.) - (k*q/(x**2 + (y+d)**2)**(1/2.)))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x,y,V, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral) # cm. then hit tab
```

```python
levels = array([8, 4, 2, 1, .5, .25, 0, -.25, -.5, -1, -2, -4, -8])
contour(x,y,V,levels,cmap=cm.coolwarm)
show()
```

\begin{equation}
G(x,y) = e^{-x^{2}-y^{2}}
\end{equation}

```python
from mpl_toolkits.mplot3d import Axes3D
k = 8.9875518*10**9 #N m^2/C^2
q1 = -1*10**-9 #C
q2 = -1*10**-9 #C
q3 = 1*10**-9 #C
q4 = 1*10**-9 #C
V1 = (k*q1/(x**2 + (y - d)**2)**(1/2.) - (k*q1/(x**2 + (y+d)**2)**(1/2.)))
V2 = (k*q2/(x**2 + (y - d)**2)**(1/2.) - (k*q2/(x**2 + (y+d)**2)**(1/2.)))
V3 = (k*q3/(x**2 + (y - d)**2)**(1/2.) - (k*q3/(x**2 + (y+d)**2)**(1/2.)))
V4 = (k*q4/(x**2 + (y - d)**2)**(1/2.) - (k*q4/(x**2 + (y+d)**2)**(1/2.)))
Vsum = V1 + V2 + V3 + V4
dx = .06
dy = .06
x,y = meshgrid(arange(-6.,6.,dx),arange(-6.,6.,dy))
x.ndim
y.shape
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x,y,Vsum, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral) # cm. then hit tab
```
