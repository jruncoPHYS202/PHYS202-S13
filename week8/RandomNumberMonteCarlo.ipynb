Example 1: Analytic Inverse CDF Transform
=

f(x) = sin(x)/2 for 0 =< x =< pi
     
     = 0 otherwise

the cumulative distribution function F(x) in this case is just the integral of f(x) below the bound:

F(x) = 0 x<0

     cos(0)-cos(x)/2 0 <= x <= pi

     1 x > pi

```python
x = arange(0.,pi+.001,.01)
f = lambda x: sin(x)/2
F = lambda x: (cos(0)-cos(x))/2

plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

F^-1(z) = cos^-1(cos(0)-2z)

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)

#sampled x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled distribution')
plot(x,f(x),label="True distribution",lw=3,color='red')
xlim(0,pi)
ylim(0,.7)
legend(loc=0)
show()
```

Exercise 1: Exponential function - analytic
=

f(x) = lambda*e^-(lambda*x)

part a/b: determine the CDF F(x) for this funtion and invert it analytically to obtain F^-1 and plot the fucntion and CDF

```python
# integral is -e^-(lambda*x)
# inverse is -log(lambdaa*x)  (ln isn't defined)
x = arange(0.,5+.001,.01)
lambdaa = 1
f = lambda x: lambdaa*exp(-lambdaa*x)
F = lambda x: 1-exp(-lambdaa*x)
invF = lambda x: -log(lambdaa*x -1)

plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)
xlim(0,5)
legend(loc=0)
show()
```

part c: generate 10000 samples according to the distribution of f(x) and plot histogram of resulting samples using 50 bins along with true distribution

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
invF = lambda x: -log(lambdaa*x)

#sampled x values
xsamples = invF(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled distribution')
plot(x,f(x),label="True distribution",lw=3,color='red')
xlim(0,5)
ylim(0,1)
legend(loc=0)
show()
```

Exercise 2: Exponential function - numerical
=

part a

```python
x = arange(0.,5.1,.1)
lambdaa = 1
f = lambda x: lambdaa*exp(-lambdaa*x)
cs = cumsum(f(x))/sum(f(x))
cs
```

part b: make the plot with 10000 samples

```python
N = 10000 #number samples
z = numpy.random.random_sample(N) #N random numbers between 0 and 1
xsamples = x[cs.searchsorted(z)] #samples x values
hist(xsamples,50,normed=True,label="sampled distribution") #compare to original
xlim(0,5.1,.1)
plot(x,f(x),label='true distribution', lw=3,color='green')
show()
```

part c: turn algorithm into function

```python
def distribute1D(x,prob,N): #prob values of f(x)
    """takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the imput data."""
    F = cumsum(prob)/sum(prob)
    z=numpy.random.random_sample(N)
    u = F.searchsorted(z)
    xsamples = x[u]
    return xsamples
b = distribute1D(x,f(x),10000)
hist(b,50,normed=True)
```
